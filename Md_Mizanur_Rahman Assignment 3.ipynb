### OpenCV2

```python
import cv2
from matplotlib import pyplot as plt # notice that we have imported pyplot differently
from google.colab.patches import cv2_imshow
```

OpenCV uses BGR image format. So, when we read an image using cv2. imread() it interprets in BGR format by default.

```python
image_path = '/content/drive/MyDrive/Image/00f2e69a-1e56-412d-8a79-fdce794a17e4___JR_B.Spot 3132.JPG'
img = cv2.imread(image_path)

cv2_imshow(img)
```

OpenCV uses BGR image format. So, when we read an image using cv2. imread() it interprets in BGR format by default. And then We show our image in Matplotlib. Which uses RGB format.

```python
image_path = '/content/drive/MyDrive/Image/00f2e69a-1e56-412d-8a79-fdce794a17e4___JR_B.Spot 3132.JPG'
img = cv2.imread(image_path)

plt.imshow(img)
```

### Color Conversion

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
cv2_imshow(img_rgb)
```
