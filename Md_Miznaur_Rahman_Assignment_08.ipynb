### What is filtering in image processing?
In computer vision filtering is a core technique used to process and analyze images by
modifying pixel values based on certain operations. Filtering can enhance important features
(like edges, corners, textures), remove noise, or transform an image to make it more suitable for
further analysis.

Applications of image filtering:
1. Noise Reduction
2. Edge Detection
3. Feature Extraction
4. Image Segmentation
5. Object Recognition

Kernel (Filter): A small matrix (e.g., 3x3, 5x5) that is applied to each pixel in the image. It represents the shape and size of the neighborhood that will be sampled when calculating the
pixel values to be modified.

Padding: When applying a filter near the edges, padding is often added to the image so that every pixel can be processed. Padding can involve filling with zeros, replicating the nearest
edge pixel, or other methods.

Convolution: A mathematical operation where the kernel slides across the image, and the pixel values are multiplied by corresponding kernel values and summed. This operation transforms
the image based on the filter applied.

Ans: Kernels in image processing are usually odd-sized (3√ó3, 5√ó5, etc.) because:

Reason: Odd-sized kernels (like 3√ó3, 5√ó5) ensure that there is a center pixel, making it easier to apply convolution operations symmetrically around a pixel.

Center Pixel Selection: An odd-sized kernel ensures that there is a well-defined center pixel. A 3√ó3 kernel has a center pixel at (1,1), a 5√ó5 kernel at (2,2), etc. This helps in proper convolution.

Symmetry Maintenance: Image filters often apply transformations symmetrically around a pixel, which is more efficient with an odd-sized kernel.

Preserves Image Structure: If an even-sized kernel is used (e.g., 2√ó2), the center is ambiguous, leading to improper filtering.

Reduces Artifacts: Odd-sized kernels help prevent artifacts in edge detection and smoothing operations.

### **Gaussian Blur: A Detailed Explanation**  

#### **What is Gaussian Blur?**  
Gaussian blur is an image processing technique used to smooth images by reducing noise and fine details. It works by applying a **Gaussian function** to the pixels in an image, weighting neighboring pixels based on their distance from the center of the kernel. This results in a more natural and smooth blurring effect compared to other techniques like box blur or median blur.  

---

### **How Gaussian Blur Works**  
1. **Convolution with a Gaussian Kernel:**  
   - A Gaussian kernel (a matrix of weights) is used to modify each pixel.
   - The value of each pixel is replaced by a **weighted average** of its neighboring pixels.
   - The Gaussian function assigns **higher weights** to pixels closer to the center and lower weights to farther pixels.

2. **Mathematical Formula of the Gaussian Function:**  
   $$ \[
   G(x, y) = \frac{1}{2\pi\sigma^2} \cdot \exp\left(\frac{- (x^2 + y^2)}{2\sigma^2}\right)
   \] $$
   - \( x, y \) ‚Üí The coordinates of a pixel relative to the kernel center.
   - \( \sigma \) ‚Üí Standard deviation of the Gaussian distribution, controlling the amount of blur.
   - A **larger** \( \sigma \) results in a **more spread-out blur**.

---

### **Why Use Gaussian Blur?**  
‚úÖ **Noise Reduction:** Helps remove unwanted noise from an image, making it smoother.  
‚úÖ **Preprocessing for Edge Detection:** Used before edge detection (like Canny edge detection) to reduce noise.  
‚úÖ **Depth of Field Simulation:** Mimics out-of-focus areas in photography.  
‚úÖ **Image Enhancement:** Softens harsh edges and makes images more aesthetically pleasing.

````python
image_load = cv2.imread('/content/terzis-fruits-orange.jpg')

# Convert to RGB (since OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image_load, cv2.COLOR_BGR2RGB)

# Filtering
kernel1 = np.ones((3, 3), np.float32)
blur_img1 = cv2.filter2D(image_rgb, -1, kernel1) # when destination depth/ddepth = -1, the output image will have the same depth as the source.

kernel2 = np.ones((7, 7), np.float32)
blur_img2 = cv2.filter2D(image_rgb, -1, kernel2)

# Gaussian Blur
G_blur_img1 = cv2.blur(image_rgb, (3, 3))

G_blur_img2 = cv2.blur(image_rgb, (7, 7))

# Print Images

fig, axis = plt.subplots(2, 3, figsize = (15, 10))

axis[0, 0].imshow(image_rgb)
axis[0, 0].set_title("Original Image")
axis[0, 0].axis("Off")

axis[0, 1].imshow(blur_img1)
axis[0, 1].set_title("Averaging 3X3 Image")
axis[0, 1].axis("Off")

axis[0, 2].imshow(blur_img2)
axis[0, 2].set_title("Averaging 7X7 Image")
axis[0, 2].axis("Off")

axis[1, 0].imshow(image_rgb)
axis[1, 0].set_title("Original Image")
axis[1, 0].axis("Off")

axis[1, 1].imshow(G_blur_img1)
axis[1, 1].set_title("Averaging 3X3 Image")
axis[1, 1].axis("Off")

axis[1, 2].imshow(G_blur_img2)
axis[1, 2].set_title("Averaging 7X7 Image")
axis[1, 2].axis("Off")

plt.show()
````
