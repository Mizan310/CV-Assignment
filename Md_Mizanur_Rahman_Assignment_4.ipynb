### 3. Take an image, apply at least 10 geometric transformations and display all of the images in a single output.

i. Flipping- here I did the 4 type of geometric transformation

```python
# Original Image
img = '/content/drive/MyDrive/Image/images.jpeg'
image = plt.imread(image_load)

horizontal_flip = cv2.flip(image, 1)
vertical_flip = cv2.flip(image, 0)
both_flip = cv2.flip(horizontal_flip, 0)
```

```python
image.shape
```

ii. Cropping

iii. Resizing, Rescaling

```python
image.shape
```

```python
plt.figure(figsize=(15,5))
plt.subplot(151)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(152)
plt.imshow(horizontal_flip)
plt.title('Horizontally flipped')

plt.subplot(153)
plt.imshow(vertical_flip)
plt.title("Vertically flipped")

plt.subplot(154)
plt.imshow(both_flip)
plt.title("Both flipped")
#########################################

(y_start, y_end, x_start, x_end) = (50, 200, 50, 200)
cropped_image = image[y_start:y_end, x_start:x_end]
plt.figure(figsize=(15,5))
plt.title("Cropped Image")
plt.imshow(cropped_image)

cropped_image.shape
#########################################

new_width = 350
new_height = 400

plt.figure(figsize=(15,5))
plt.subplot(131)
resized_img = cv2.resize(cropped_image, (new_width, new_height))
plt.title("Resizing")
plt.imshow(resized_img)

plt.subplot(132)
resize_img = cv2.resize(cropped_image, None, fx=.5, fy=.5, interpolation = cv2.INTER_CUBIC)
plt.title("Rescaling")
plt.imshow(resize_img)

#############################

rotated_image = cv2.rotate(image, cv2.ROTATE_180)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated Image')

###########################################

import cv2
(h, w) = image.shape[:2]

center = (w // 2, h // 2)

M1 = cv2.getRotationMatrix2D(center, 30, 1.0) # (center: Point2f, angle: float, scale: float)
M2 = cv2.getRotationMatrix2D(center, 45, 1.0) # (center: Point2f, angle: float, scale: float)
M3 = cv2.getRotationMatrix2D(center, -90, 1) # (center: Point2f, angle: float, scale: float)
M4 = cv2.getRotationMatrix2D(center, 90, 1.0) # (center: Point2f, angle: float, scale: float)

rotated_image_30 = cv2.warpAffine(image, M1, (w, h))
rotated_image_45 = cv2.warpAffine(image, M2, (w, h))
rotated_image_60 = cv2.warpAffine(image, M3, (w, h))
rotated_image_90 = cv2.warpAffine(image, M4, (w, h))

plt.figure(figsize=(15, 5))

plt.subplot(141)
plt.imshow(rotated_image_30)
plt.title('30 degree roated')

plt.subplot(142)
plt.imshow(rotated_image_45)
plt.title('45 degree roated')

plt.subplot(143)
plt.imshow(rotated_image_60)
plt.title('60 degree roated')

plt.subplot(144)
plt.imshow(rotated_image_90)
plt.title('90 degree roated')

plt.tight_layout()
plt.show()
```
